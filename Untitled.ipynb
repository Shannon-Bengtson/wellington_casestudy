```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cartopy
import datetime
import pyproj
import rioxarray
```

```python
# Define the current and target CRS using EPSG codes
current_crs = pyproj.CRS("EPSG:2193")  # For example, this is WGS 84
target_crs = pyproj.CRS("EPSG:4326")
transformer = pyproj.Transformer.from_crs(current_crs, target_crs, always_xy=True)
```

```python
# Create a regular grid of coordinates
x_regular = np.linspace(xr_dem['x'].min(), xr_dem['x'].max(), np.max([len(xr_dem['x']),len(xr_dem['y'])]))
y_regular = np.linspace(xr_dem['y'].min(), xr_dem['y'].max(), np.max([len(xr_dem['x']),len(xr_dem['y'])]))

# Interpolate data onto regular grid
data_interpolated = xr_dem.band_data[0,:,:].interp(y=y_regular, x=x_regular)

# Now you can use the interpolated data for transformation
new_DEM = xr.DataArray(data_interpolated,coords={"y":y_regular,"x":x_regular},dims=["y","x"])

xx,yy = transformer.transform(new_DEM.x,new_DEM.y)

new_DEM['x'] = xx
new_DEM['y'] = yy
```
